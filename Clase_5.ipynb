```python
%%writefile Suma.java
// ==========================================
// EJERCICIO 2: CLASE "Suma"
// Objetivo: explicar métodos con parámetros y retorno
// ==========================================

// Clase que realiza una suma muy simple
class Suma {

    // Método que recibe 2 números y retorna la suma
    int sumar(int a, int b) {
        // Explicación:
        // - 'a' y 'b' son parámetros del método
        // - el método retorna a + b
        return a + b;
    }
}

// Clase principal que prueba el método sumar()
class SumaTest {
    public static void main(String[] args) {

        // 1) Crear un objeto de la clase Suma
        Suma calculadora = new Suma();

        // 2) Llamar al método sumar() enviando 2 números
        int resultado = calculadora.sumar(5, 3);

        // 3) Mostrar por pantalla el resultado
        System.out.println("La suma de 5 + 3 es: " + resultado);

        // ¿QUÉ HACE ESTE CÓDIGO?
        // - Crea un objeto llamado calculadora
        // - Llama al método sumar(5,3)
        // - El método retorna 8
        // - Se imprime ese valor
    }
}
```

```python
%%writefile Banco.java
import java.util.Scanner;

// 1) Clase Banco (modelo de una cuenta bancaria)
class Banco {
    // Atributos privados (encapsulamiento)
    private String nombre;
    private double saldo;

    // 1.1 Constructor: inicializa nombre y saldo
    public Banco(String nombre, double saldoInicial) {
        this.nombre = nombre;
        this.saldo = saldoInicial;
    }

    // 1.2 Método para depositar dinero
    public void depositar(double monto) {
        if (monto > 0) {
            saldo += monto;
            System.out.println("Depósito exitoso. Nuevo saldo: $" + saldo);
        } else {
            System.out.println("El monto debe ser positivo.");
        }
    }

    // 1.3 Método para retirar dinero
    public void retirar(double monto) {
        if (monto > saldo) {
            System.out.println("Fondos insuficientes. Saldo actual: $" + saldo);
        } else if (monto <= 0) {
            System.out.println("El monto debe ser mayor que cero.");
        } else {
            saldo -= monto;
            System.out.println("Retiro exitoso. Nuevo saldo: $" + saldo);
        }
    }

    // 1.4 Método para mostrar información de la cuenta
    public void mostrarInformacion() {
        System.out.println("Banco: " + nombre);
        System.out.println("Saldo actual: $" + saldo);
    }

    // 1.5 Getter para leer el saldo de manera segura
    public double getSaldo() {
        return saldo;
    }

    // 1.6 Setter para cambiar el nombre del banco
    public void setNombre(String nuevoNombre) {
        nombre = nuevoNombre;
    }
}

// 2) Clase BancoTest (tiene el método main y el menú)
// OJO: aquí ya NO es public
class BancoTest {
    public static void main(String[] args) {

        // Creamos el objeto Scanner para leer desde teclado
        Scanner scanner = new Scanner(System.in);

        // Creamos un objeto Banco con nombre y saldo inicial
        Banco cuenta = new Banco("Banco Ejemplo", 1000);

        int opcion; // variable para guardar la opción del menú

        // Ciclo do-while para mostrar el menú hasta que el usuario elija salir
        do {
            System.out.println("\n--- MENÚ CUENTA BANCARIA ---");
            System.out.println("1. Ver información");
            System.out.println("2. Depositar");
            System.out.println("3. Retirar");
            System.out.println("4. Cambiar nombre del banco");
            System.out.println("0. Salir");
            System.out.print("Ingrese opción: ");

            opcion = scanner.nextInt(); // leemos la opción

            switch (opcion) {
                case 1:
                    cuenta.mostrarInformacion();
                    break;

                case 2:
                    System.out.print("Ingrese monto a depositar: ");
                    double montoDeposito = scanner.nextDouble();
                    cuenta.depositar(montoDeposito);
                    break;

                case 3:
                    System.out.print("Ingrese monto a retirar: ");
                    double montoRetiro = scanner.nextDouble();
                    cuenta.retirar(montoRetiro);
                    break;

                case 4:
                    System.out.print("Ingrese nuevo nombre del banco: ");
                    scanner.nextLine(); // limpiar buffer
                    String nuevoNombre = scanner.nextLine();
                    cuenta.setNombre(nuevoNombre);
                    System.out.println("Nombre del banco actualizado.");
                    break;

                case 0:
                    System.out.println("Saliendo del sistema. ¡Gracias!");
                    break;

                default:
                    System.out.println("Opción no válida. Intente nuevamente.");
                    break;
            }

        } while (opcion != 0);

        scanner.close();
    }
}
```
